```python
electron_runlist = hgc.runlist.query("Particle == 'electron'").query('Configuration == "23c"').query('Energy == "300"')
electron_runlist
```

```python
rechit_energy_mean = []
rechit_energy_variance = []
rechit_x_mean = []
rechit_x_variance = []
rechit_y_mean = []
rechit_y_variance = []
rechit_z_mean = []
rechit_z_variance = []
configuration = []
energy = []
```

```python
run_numbers = electron_runlist.Run.values
Config = electron_runlist.Configuration
for (run,Configuration) in zip (run_numbers,Config):
    hdf = hgc.load_run(run, columns = ["beamEnergy","event", "rechit_layer", "rechit_energy", "rechit_x", "rechit_y", "rechit_z"])
    configuration.append(Configuration)
    energy.append(hdf.beamEnergy[0][0])
    rechit_energy_mean.append(hdf.rechit_energy.mean())
    rechit_energy_variance.append(hdf.rechit_energy.std())
    rechit_x_mean.append(hdf.rechit_x.mean())
    rechit_x_variance.append(hdf.rechit_x.std())
    rechit_y_mean.append(hdf.rechit_x.mean())
    rechit_y_variance.append(hdf.rechit_x.std())
    rechit_z_mean.append(hdf.rechit_x.mean())
    rechit_z_variance.append(hdf.rechit_x.std())

Type = pd.DataFrame({
    "energy_mean" : rechit_energy_mean,
    "energy_variance" : rechit_energy_variance,
    "x_mean" : rechit_x_mean,
    "x_variance" : rechit_x_variance,
    "y_mean" : rechit_y_mean,
    "y_variance" : rechit_y_variance,
    "z_mean" : rechit_z_mean,
    "z_variance" : rechit_z_variance,
    "configuration" : configuration,
    "Beam_energy" : energy    
})
Type

```
